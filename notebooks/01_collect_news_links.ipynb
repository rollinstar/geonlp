```python
from modules.news_collector import DaumNewsCollector
import pandas as pd
import json
```

```python
dnc = DaumNewsCollector(1.5, 5)
```

```python
# collect 20190101 ~ 20190131
links_2019_Jan = []
for i in range(1, 32):
    day = str(i)
    if i < 10:
        day = day.zfill(2)
    d = '201901' + day
    print('Start > ', d)
    links_2019_Jan += dnc.create_news_pages_links_by_date(d)
    print('End > ', d)
```
