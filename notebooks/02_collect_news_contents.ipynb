```python
from modules.news_collector import DaumNewsCollector
import json
import time
from random import uniform
import pandas as pd
```

```python
with open('./data/source/links_201901.json', 'r') as f:
    links = json.load(f)
```

```python
links[0]
```

```python
# 1 ~ 5sec interval
dnc = DaumNewsCollector(1, 5) 
```

```python
c = dnc.create_news_contents_data('20190101235354271')
c
```

```python
news_contents = []
for link in links:
    news_ids = link['news_ids']
    day = link['news_page_info']
    print('Start day > ', day)
    t = []
    for nid in news_ids:
        print(nid)
        content = dnc.create_news_contents_data(nid)
        news_contents.append(content)
        f = uniform(1, 5)
        time.sleep(f)
        t.append(str('%.2f' % f) + 'sec')
    print('time > ', t)
    print('End day > ', day)
```

```python
len(news_contents)
```

```python
df = pd.DataFrame.from_dict(news_contents)
df.head(5)
```

```python
with open('./data/news/news_201901_01.json', 'w', encoding='utf-8') as make_file:
    json.dump(news_contents, make_file, indent="\t")
```
